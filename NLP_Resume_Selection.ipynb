```python
resume=docx2txt.process('Resume.docx')

```

```python
profile=docx2txt.process('Job_profile.docx')
```

```python
text=[resume,profile]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
matrix=cv.fit_transform(text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(matrix))
```

```python
percentage=cosine_similarity(matrix)[0][1]*100
percentage=round(percentage,2)
print('Your resume matches is about '+str(percentage)+ '% of job profile')
```
